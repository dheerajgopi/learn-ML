```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('dataset/kaggle-house-price/train.csv')
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
sns.distplot(df['SalePrice'], kde=True)
```

## Numerical features

```python
corr = df.select_dtypes(include=['float64', 'int64']).iloc[:, 1:].corr()
plt.figure(figsize=[12, 12])
sns.heatmap(corr, square=True);
```

```python
saleprice_corr = corr['SalePrice'].drop('SalePrice')
print "List features in descending order of their correlation with saleprice"
saleprice_corr.sort_values(ascending=False)
```

The sale price correlates strongly with **OverallQual, GrLivArea, GarageCars, GarageArea, TotalBsmtSF, 1stFlrSF, FullBath, TotRmsAbvGrd, YearBuilt, YearRemodAdd, GargeYrBlt, MasVnrArea and Fireplaces**.

```python
sns.regplot(x=df['OverallQual'], y=df['SalePrice']);
```

```python
plt.figure(1)
f, axarr = plt.subplots(3, 2, figsize=(10, 9))
price = df['SalePrice']
axarr[0, 0].scatter(df['GrLivArea'], price)
axarr[0, 0].set_title('GrLivArea')
axarr[0, 1].scatter(df['GarageArea'], price)
axarr[0, 1].set_title('GarageArea')
axarr[1, 0].scatter(df['TotalBsmtSF'], price)
axarr[1, 0].set_title('TotalBsmtSF')
axarr[1, 1].scatter(df['1stFlrSF'], price)
axarr[1, 1].set_title('1stFlrSF')
axarr[2, 0].scatter(df['TotRmsAbvGrd'], price)
axarr[2, 0].set_title('TotRmsAbvGrd')
axarr[2, 1].scatter(df['MasVnrArea'], price)
axarr[2, 1].set_title('MasVnrArea')
f.text(-0.01, 0.5, 'Sale Price', va='center', rotation='vertical', fontsize = 12)
plt.tight_layout()
plt.show();
```
