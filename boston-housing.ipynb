```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets

%matplotlib inline
```

```python
# load dataset
boston = datasets.load_boston()
```

```python
# target variable (value of house / 1000 ($))
# 'boston' is not a pandas DataFrame, so use 'reshape' to form a column
yb = boston.target.reshape(-1, 1);
```

```python
# scatter plot of each predictor variables vs target variable (value of house / 1000 ($))
# this will reveal the correlation between each predictor variables and the target variable
num_features = np.shape(boston.data)[1]
lin_reg = LinearRegression()

# plot settings
plt.style.use('ggplot')
plt.figure(figsize=[20, 25])
num_cols = 3
num_rows = (num_features / num_cols) + (num_features % num_cols)

for num in range(num_features):
    # train the model using one predictor variable at a time
    Xb = boston.data[:, num].reshape(-1, 1) # reshape each predictor var into a column
    plt.subplot(num_rows, num_cols, num + 1)
    feature_name = boston.feature_names[num]
    plt.scatter(Xb, yb, color='black')
    plt.xlabel(feature_name)
    plt.ylabel('value of house / 1000 ($)');
    lin_reg.fit(Xb, yb)
    plt.plot(Xb, lin_reg.predict(Xb), color='blue', linewidth=3)

plt.show()
```
